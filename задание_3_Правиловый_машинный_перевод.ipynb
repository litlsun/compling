## 3. Приведите 3 примера правил для этих языков

**Правило 1: Перевод падежей**

*В греческом языке падежи выражаются через окончания существительных и артикли, что аналогично падежным окончаниям в русском языке.*

1. Определить падеж существительного в русском предложении: вижу (кого/что?) собаку → винительный падеж
2. Найти соответствующий артикль и окончание в греческом языке. "τον" — артикль мужского рода в винительном падеже
3. Перевести существительное с учетом падежа: "собаку" → "τον σκύλο".
4. Результат: "Я вижу собаку." → "Βλέπω τον σκύλο."

**Правило 2: Времена глаголов**

*В греческом языке время глагола также выражается через спряжение, как и в русском, но с использованием различных окончаний и вспомогательных слов.*

1. Определить время глагола в русском предложении: буду читать - будущее время
2. Найти эквивалентное время и форму глагола в греческом языке. "Θα" — вспомогательное слово для будущего времени
3. Проспрягать глагол согласно лицу и числу: "буду читать" → "θα διαβάσω".
4. Результат: "Я буду читать книгу." → "Θα διαβάσω το βιβλίο."

**Правило 3: Использование предлогов**

*Предлоги в греческом языке могут иметь разные значения по сравнению с русскими предлогами.*

1. Определить предлог в русском предложении. (предлог места "в")
2. Найти эквивалентный предлог в греческом языке, учитывая контекст. στην - предлог места, аналогичный русскому "в"
3. Результат: "Я живу в городе." → "Ζω στην πόλη."
